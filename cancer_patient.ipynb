```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df = pd.read_csv("Resources/cancerpatientdata.csv")

# Review the DataFrame
df.head()
```

```python
# df2 = df.replace({'Level': 'Low', 'Level': 'Medium', 'Level': 'High'},
# {'Level': 1, 'Level': 2, 'Level': 3}, regex=True)
# df2 = df.replace('Medium','2', regex=True)
# df2 = df.replace('High','3', regex=True)
df['Level'] = df['Level'].replace(['Low', 'Medium', 'High'], [1, 2, 3])
df
```

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = df.Level

# Separate the X variable, the features
X = df.drop(columns='Level')
X = df.drop(columns='Patient Id')
```

```python
# Review the y variable Series
y.head()
```

```python
# Review the X variable DataFrame
X.head()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
classifier = LogisticRegression(random_state=1)
classifier

# Fit the model using training data
classifier.fit(X_train, y_train)
```

```python
 # Score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
 # Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
 from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

```python
# Generate a confusion matrix for the model
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
# Print the classification report for the model
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
# Count the distinct values of the resampled labels data
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_train,y_train)
```

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier1 = LogisticRegression(random_state=1)
classifier1

# Fit the model using the resampled training data
classifier1.fit(X_resampled, y_resampled)

 # Score the model
print(f"Training Data Score: {classifier1.score(X_resampled, y_resampled)}")
print(f"Testing Data Score: {classifier1.score(X_test, y_test)}")

# Make a prediction using the testing data
predictions = classifier1.predict(X_resampled)
pd.DataFrame({"Prediction": predictions, "Actual": y_resampled})
```

```python
# Print the balanced_accuracy score of the model 
accuracy_score(y_resampled, predictions)
```

```python
# Generate a confusion matrix for the model
confusion_matrix(y_resampled, predictions)
```

```python
# Print the classification report for the model
print(classification_report(y_resampled, predictions))
```
